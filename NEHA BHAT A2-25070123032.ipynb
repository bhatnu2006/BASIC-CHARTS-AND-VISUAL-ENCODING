```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Days':['Mon','Tue','Wed','Thurs','Fri'],
    'Study_Hours':[1,2,6,4,5],
    'Marks':[89,78,90,99,98],
    'Attendance':[75,78,89,85,89],
    'Sleep_Hours':[8,7,6,8,3],
    'Assignment_Completed':[1,3,2,4,1]
}
df=pd.DataFrame(data)
print(df)
```

```python
# 1. LINE CHARTS
plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.show()
```

```python
# 1.2. LINE CHARTS
plt.plot(df['Days'],df['Study_Hours'],marker='o')
plt.title('Study hours trend')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.show()
```

```python
# ADVANCED LINE CHART
plt.figure(figsize=(7,4))
plt.plot(df['Days'],df['Study_Hours'],
         marker='o',color='blue',label='Study_Hours')
plt.plot(df['Days'],df['Marks'],
         marker='s',color='red',label='Marks')
plt.title('Study hours and Marks Trend')
plt.xlabel('Days')
plt.ylabel('Values')
plt.legend()
plt
```

```python
# 2. BAR CHART
plt.bar(df['Days'],df['Marks'], color='pink')
plt.title('Marks Comparison')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
# ADVANCED BAR CHARTS
bars = plt.bar(df['Days'],df['Marks'],color='pink')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,y,str(y),
             ha='center',va='bottom')

plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks Comparison')
plt.grid(axis='y')
plt.show()
```

```python
# 4.HISTOGRAM
data ={
    'Marks':[70,80,90,98,78,87,67,65,78,89,96,93,92,91,89]
}
df=pd.DataFrame(data)
plt.hist(df['Marks'],bins=5,color='lavender')
plt.title('Marks Distribution')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt
```

```python
#4. SCATTER PLOT
plt.scatter(df['Study_Hours'],df['Marks'])
plt.title('Study_Hours vs Marks')
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# CONDITIONAL SCATTER PLOT

# Assign 'Result' column correctly
df['Result']=['Fail','Pass','Fail','Pass','Pass']

# Fix SyntaxError in list comprehension
colors = ['red' if r == 'Fail' else 'green' for r in df['Result']]

plt.scatter(df['Study_Hours'],df['Marks'],c=colors)
plt.title('Study_Hours vs Marks')
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.show()
```

## GRAPHS USING SEABORN
